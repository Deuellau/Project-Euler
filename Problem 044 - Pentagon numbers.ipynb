### Problem 44 - Pentagon numbers

Pentagonal numbers are generated by the formula, Pn=n(3n−1)/2. The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that P4 + P7 = 22 + 70 = 92 = P8. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, Pj and Pk, for which their sum and difference are pentagonal and D = |Pk − Pj| is minimised; what is the value of D?

```python
def pentagonalnumbers():
    n = 1
    while True:
        yield n * (3 * n - 1) // 2
        n += 1

def problemFortyFour(number):
    pentalst = []
    is_penta = set()

    for p in pentagonalnumbers():
        if p > number:
            break
        pentalst.append(p)
        is_penta.add(p)

    min_diff = float("inf")
    for i in range(len(pentalst)):
        for j in range(i+1, len(pentalst)):
            penta_i, penta_j = pentalst[i], pentalst[j]
            if penta_i - penta_j >= min_diff:
                break
            if penta_i + penta_j in is_penta and penta_j - penta_i in is_penta:
                min_diff = penta_j - penta_i
                break

    return min_diff


problemFortyFour(10000000)
```
